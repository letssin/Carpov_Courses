```python
bookings['total_kids'] = bookings.children + bookings.babies
```

```python
bookings.groupby('hotel').aggregate({'total_kids':'mean'})
```

```python
bookings['has_kids'] = False
```

```python
for index in bookings.index:
    if bookings.iloc[index,-2] > 0:
        bookings.iloc[index,-1] = True
```

```python
bookings.shape
```

```python
by_kids_canc = bookings.groupby(['has_kids' ,'is_canceled'], as_index = False)\
    .agg({'customer_type':'count'})
```

```python
by_kids_canc
```

```python
by_kids = bookings.groupby(['has_kids'])\
    .agg({'customer_type':'count'})\
    .rename(columns = {'customer_type':'Total'})
```

```python
total = by_kids_canc.merge(by_kids, on = 'has_kids', how = 'inner')
```

```python
total['Perc'] = (total['customer_type'] / total['Total']).mul(100)
```
