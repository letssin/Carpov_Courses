```python
df = pd.DataFrame(columns = ['date','name','product_id','quantity'])
```

```python
df = pd.DataFrame()
```

```python
for dirr in os.listdir(root):
    for fold in os.listdir(root + '/'+ dirr):
        for file in os.listdir(root + '/'+ dirr + '/'+ fold):
            temp_df = pd.read_csv(root + '/'+ dirr + '/'+ fold + '/data.csv')
            temp_df['date'] = dirr
            temp_df['name'] = fold
            df = pd.concat([df, temp_df], ignore_index=True).drop(['Unnamed: 0'], axis = 1)
            
    
    
    
```

```python
df.head(20)
```

```python
df.quantity.sum()
```

```python
df.groupby('name')\
    .agg({'quantity':'sum'})\
    .sort_values('quantity', ascending = False)
```

```python
prod = df.groupby('product_id', as_index = False)\
    .agg({'quantity':'sum'})\
    .sort_values('quantity', ascending = False)\
    .head(10)\
    .reset_index()
```

```python
by_day = df.groupby('date', as_index = False)\
    .agg({'quantity':'sum'})
    
```

```python
sns.barplot(data = prod, x = 'product_id', y = 'quantity', color = 'green', palette = 'pastel')
```
