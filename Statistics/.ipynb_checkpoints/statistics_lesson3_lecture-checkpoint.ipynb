```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
mu = 100
sigma = 10
```

```python
#представим, что это сколько наши пользователи тратят рублей на сайте. Генеральная совоупность
population = np.random.normal(mu, sigma, 10000)
```

```python
sns.displot(population, kde = False)
```

```python
mu = population.mean()
```

```python
sigma = population.std()
```

```python
sample_size = 30
```

```python
sample = np.random.choice(population, sample_size, False)
```

```python
se = sigma / sample_size ** 0.5
```

```python
z_scores=[]
```

```python
n = 10000
for i in range(n):
    sample = np.random.choice(population, sample_size, False)
    sample_mean = sample.mean()
    
    z = (sample_mean - mu) / se
    z_scores.append(z)
```

```python
sns.displot(z_scores, kde = False)
```

```python
t_scores = []
```

```python
n = 1000
for i in range(n):
    sample = np.random.choice(population, sample_size, False)
    sample_mean = sample.mean()
    sample_std = sample.std()
    se_ = sample_std / sample_size ** 0.5
    
    t = (sample_mean - mu) / se_ #оставляем mu потому что тестируем гипотезу о среднем в ГС
    t_scores.append(t)
```

```python
sns.displot(t_scores, kde = False)
```

не можем сказать чему точно равно среднее в ГС, но можем задать границы в которых среднее скорее всего лежит

```python
sample_means = []
```

```python
n = 1000
```

```python
for i in range(n):
    sample = np.random.choice(population, sample_size, False)
    sample_means.append(sample.mean())
```

```python
sns.displot(sample_means, kde = True)
```

Правило 3х сигм ~ 100% всех наблюдений леижт в пределах 3х сигм<br>
Правило 1,95 сигм ~ 95% всех наблюдений леижт в пределах этого диапазона<br>
se = sigma / корень(n)

```python
se = population.std() / sample_size **0.5
se
```

```python
np.std(sample_means)
```

```python
population.mean() - 1.96*se
```

```python
population.mean() + 1.96*se
```
