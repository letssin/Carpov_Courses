```python
import pandas as pd
import numpy as np

import scipy.stats as ss
import statsmodels.api as sm
import statsmodels.formula.api as smf

from statsmodels.stats.api import anova_lm
from statsmodels.stats.multicomp import (pairwise_tukeyhsd, MultiComparison)

import pingouin as pg
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
post_likes = pd.read_csv('post_likes.csv')
```

```python
post_likes.head()
```

```python
sns.boxplot(data = post_likes, x = 'button', y = 'likes' , palette = ['r','g','b'])
```

```python
sns.violinplot(data = post_likes, x = 'button', y = 'likes' , palette = ['r','g','b'])
```

#### Три способа посчитать однофакторный дисперсионный анализ

```python
red = post_likes.query("button == 'red'").likes
green = post_likes.query("button == 'green'").likes
blue = post_likes.query("button == 'blue'").likes
```

```python
ss.f_oneway(red, green, blue)
```

```python
#через statsmodels

model = smf.ols(formula = "likes ~ C(button)", data = post_likes).fit()
anova_lm(model)
```

likes - переменная по которой ищем различия<br>
C(button) - независимая переменная. Обозначение групп<br>
С() - обозначает категориальную переменную

```python
smf.ols(formula = "likes ~ C(button)", data = post_likes).fit()
```

```python
#Через pingouin
pg.anova(data = post_likes, dv = 'likes', between = 'button')
```

```python
#Среднее с доверительными интервалами
sns.pointplot(data = post_likes , x = 'button', y = 'likes');
```

#### Тестируем нормальность

```python
# Через scipy
# Метод Шапиро-Уилка
# Тестируем нулевую гипотезу, что данные распределены нормально
print(ss.shapiro(red))
print(ss.shapiro(blue))
print(ss.shapiro(green))
```

```python
# scipy.normaltest
print(ss.normaltest(red))
print(ss.normaltest(blue))
print(ss.normaltest(green))
```

```python
pg.normality(data = post_likes, dv = 'likes', group = 'button', method = 'normaltest')
```

Тест Шапиро-Уилка не очень хорошо реагирует на выборки очень большого размера<br>
Normaltest более адекватно реагирует на большие выборки

#### Многофакторный ANOVA

```python
ads_clicks = pd.read_csv('ads_clicks.csv')
```

```python
ads_clicks.head()
```

```python
# Через statmodels, только главные эффкекты

model = smf.ols("clicks ~ C(ads) + C(age_group)", data = ads_clicks).fit()
anova_lm(model)
```

Мы получили то, что называется Главными эффектами. Это то, как влияют рекламы и возрастные группы не зависимо друг от друга, т.е. предполагаем, что взаимодейситвия между факторами нет

```python
# Один фактор
sns.pointplot(data = ads_clicks, x = 'ads', y = 'clicks');
```

```python
# Второй фактор
sns.pointplot(data = ads_clicks, x = 'age_group', y = 'clicks');
```

```python
# Хотим проверить взаимодействие

model_interact = smf.ols("clicks ~ C(ads) + C(age_group) + C(ads):C(age_group)", data = ads_clicks).fit()
anova_lm(model_interact)
```

```python
sns.pointplot(data = ads_clicks, x = 'ads', y = 'clicks', hue = 'age_group');
```

```python
#Через pingouin
pg.anova(data = ads_clicks, dv = 'clicks', between = ['age_group','ads'])
```

```python
# Другой размер еффекта
pg.anova(data = ads_clicks, dv = 'clicks', between = ['age_group','ads'], effsize='n2')
```

```python
pg.anova(data = ads_clicks, dv = 'clicks', between = ['age_group','ads'], ss_type= 3)
```
