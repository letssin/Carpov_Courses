```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
london = pd.read_csv('london.csv', parse_dates=['timestamp'])
```

```python
london.head()
```

```python
london.dtypes
```

```python
london.isnull().sum()
```

```python

sns.lineplot(data=london, x="timestamp", y="cnt")
```

```python
resampled = london.resample(rule='D', on='timestamp').sum()
```

```python
sns.lineplot(data=resampled, x="timestamp", y="cnt")
```
