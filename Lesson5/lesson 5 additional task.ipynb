```python
path = 'shared/homeworks/python_ds_miniprojects/5_subsid/subsid/'
```

```python
sales_df = pd.DataFrame()
for file in os.listdir(path)[1::]:
    temp_df = pd.read_csv(path + file, sep = ';')
    temp_df = temp_df[sorted(temp_df.columns)]
    sales_df = pd.concat([sales_df, temp_df], ignore_index=True)
    sales_df = sales_df.dropna()
```

```python
sales_df['SUBS_ID'] = sales_df.SUBS_ID.apply(lambda x: 'id' + x if not x.startswith('id') else x)
```

```python
sales_df.head(20)
```

```python
logs = pd.read_csv(path + 'prod_activations_logs.csv', sep = ';')
```

```python
logs.head()
```

```python
logs.dtypes
```

```python
logs['START_DTTM'] = pd.to_datetime(logs.START_DTTM, format='%d-%m-%Y %H:%M')
logs['END_DTTM'] = pd.to_datetime(logs.END_DTTM, format='%d-%m-%Y %H:%M')
```

```python
logs['difference '] = logs['END_DTTM'] - logs['START_DTTM']
```

```python
logs['minutes'] = logs['difference '] / pd.Timedelta(minutes=1)
```

```python
logs['ok_sales'] = logs.minutes.apply(lambda x: True if x >= 5 else False)
```

```python
 final_df = logs.query('ok_sales == True').merge(sales_df, how='inner')
```
