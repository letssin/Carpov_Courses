```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

<b>Figure</b> - риснуок является общим окном в котором можно построить неск графиков <b>axes</b>

```python
fig, ax = plt.subplots(2,2)
# fig - рисунок, axes - матрица 2 на 2 (из 4х графиков)
```

```python
#добавляем графики по отдельности
figure = plt.figure()

ax1 = figure.add_subplot(2,2,1)  #расчерчивем рисунок на 4 части, вставляем график в 1ю ячейку
ax2 = figure.add_subplot(2,2,2)  #вставляем график в 2ю ячейку
ax3 = figure.add_subplot(2,2,3)  #вставляем график в 3ю ячейку
```

```python
#2 графика на 1 картинке
fig, ax = plt.subplots()
x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]
x_abs = [3, 2, 1, 0, 1, 2, 3]
ax.plot(x,y)
ax.plot(x, x_abs);
```

```python
fig, axes = plt.subplots(2,2)
x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]
x_abs = [3, 2, 1, 0, 1, 2, 3]
axes[1,0].plot(x,y) # передаем график в элемент матрицы с номером [1,0] - 3я ячейка
axes[0,0].plot(x,x_abs) # передаем график в элемент матрицы с номером [1,0] - 3я ячейка
```

```python
figure = plt.figure()
ax1 = figure.add_subplot(2,2,1)
ax2 = figure.add_subplot(2,2,2)
ax3 = figure.add_subplot(2,2,3)

ax1.plot(x,y)
ax3.plot(x,x_abs)
```

### Цвет линий и стили

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x,y, color = 'green', linestyle = ':', marker = 'o') 

ax.plot(x,np.sin(x), color = 'red', linestyle = '-.', marker = '*')

ax.plot(x,np.cos(x), color = '0.75', linestyle = '--', marker = '+') # color = '0' белый 1 черный
```

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x,y, color = 'red', linestyle = '--', marker = '*', alpha = 0.2); #alpha - прозрачность 0-прозрачно(не видно)
ax.plot(0.5*x, y, 'o:g', linewidth = 2.5) # 'o:g' строка с параметрами о-маркер :- линия g- цвет
```

### Лимиты осей

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x,y, color = 'red', marker = '*')
#лимиты по осям
ax.set_xlim(0,11)
ax.set_ylim(-5,10)
```

### Подписи и метки

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x,y, label = 'x^2')
ax.plot(x*0.5, y, color = 'blue',label = '0.5x^2')

ax.set_title('Заголовок графика', fontsize = 20)

ax.set_xlabel('ось X')
ax.set_ylabel('ось Y')

ax.legend(loc = 'lower left');
```

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0, 1, 2, 3])
y = np.array([9, 4, 1, 0, 1, 4, 9])

ax.plot(x,y, label = 'x^2', marker = 'o')
ax.legend(loc = 'lower left')

ax.set_xticks(np.arange(-3,3,0.5)) #tikcs - подписи по оси 
ax.set_yticks(np.arange(0,10,3))

ax.spines['right'].set_visible(False) #нет границы справа

ax.grid(color = 'grey', linewidth = 1.5, linestyle = '--') # настройка сетки
# ax.grid() - убрать сетку
;
```

### Сохраняем и открываем картинку

```python
fig.savefig('figure1.png')

import matplotlib.image as mpimg

img = mpimg.imread('figure1.png')
plt.grid()
plt.imshow(img)
```

## Различные виды графиков
#### Scatteplot

```python
fix, ax = plt.subplots()

N = 50

x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)

area = (30 * np.random.rand(N)) ** 2

ax.scatter(x, y, s = area, c = colors, alpha = 0.5, marker = '*')
```

#### Barplot

```python
fig, ax = plt.subplots(1,2)
ax[0].bar([1, 2, 3], [3, 4, 5], color = 'red', label = 'vert bars', alpha = 0.5)
ax[1].barh([0.5, 1, 2.5], [0,1,2]) #barh горизонтальные бары
ax[0].legend()
```
